Artificial neural networks (ANNs) are computing systems vaguely inspired by the biological neural networks that constitute animal brains. At their core, ANNs are composed of layers of interconnected nodes, similar to neurons, which can process input data and generate output. The connections between these nodes can be adjusted through a process called learning, which improves the networkâ€™s performance over time.

### Basic Components of an ANN
1. **Input Layer:** Receives various forms of raw data.
2. **Hidden Layers:** Intermediate layers that process inputs received from the input layer using weighted connections. These layers can have various architectures to perform different transformations on the input.
3. **Output Layer:** Produces the final output, which can be a classification or regression value.

### Learning in Neural Networks
Neural networks learn through an algorithm called backpropagation, which involves:
- **Forward Pass:** Input data is passed through the network, layer by layer, until the output layer is reached.
- **Loss Calculation:** The output is compared to the true value, and the error (or loss) is calculated.
- **Backward Pass (Backpropagation):** The error is sent back through the network, adjusting the weights of the connections to minimize the error.


For a practical example, let's consider a dataset often used for public health studies: the **Heart Disease UCI dataset** available from the UCI Machine Learning Repository. This dataset is widely used in machine learning to predict whether a patient has heart disease based on various medical attributes.

### Dataset Overview:
The dataset includes several attributes such as:
- Age: Age of the patient
- Sex: Sex of the patient
- CP (Chest Pain type): Type of chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)
- Trestbps (Resting Blood Pressure): In mm Hg on admission to the hospital
- Chol (Serum Cholestrol): In mg/dl
- Fbs (Fasting Blood Sugar): If > 120 mg/dl (1 = true; 0 = false)
- RestECG (Resting Electrocardiographic results): Values 0,1,2
- Thalach (Maximum Heart Rate achieved)
- Exang (Exercise Induced Angina): 1 = yes; 0 = no
- Oldpeak (ST depression induced by exercise relative to rest)
- Slope (The slope of the peak exercise ST segment)
- CA (Number of major vessels colored by fluoroscopy): 0-3
- Thal: 3 = normal; 6 = fixed defect; 7 = reversible defect
- Target: Diagnosis of heart disease (1 = present, 0 = not present)

### Sample Python Implementation:
We'll use `scikit-learn` for data preprocessing and model creation, and `keras` from TensorFlow for building a simple artificial neural network.


#### 1. Load and Prepare Data


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load dataset
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
data = pd.read_csv(url, names=column_names, na_values="?")

# Fill missing values
data.fillna(data.mean(), inplace=True)

# Split data into features and labels
X = data.drop('target', axis=1)
y = data['target']

# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

#### 2. Build and Train ANN Model


```python
from keras.models import Sequential
from keras.layers import Dense

# Create model
model = Sequential()
model.add(Dense(12, input_dim=13, activation='relu'))  # Input layer + hidden layer with 12 neurons
model.add(Dense(8, activation='relu'))                 # Another hidden layer with 8 neurons
model.add(Dense(1, activation='sigmoid'))              # Output layer

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(X_train, y_train, epochs=150, batch_size=10, verbose=1)
```

```python
# Evaluate the model
_, accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy: {accuracy*100:.2f}%")
```
